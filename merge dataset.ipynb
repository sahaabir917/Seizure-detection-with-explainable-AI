```python
#  Write code for training dataset 
import h5py
import pandas as pd
import numpy as np


h5_path = "newdataset.hdf5"
csv_path = "newdataset.csv"


```

```python
#Check the data shape

with h5py.File(h5_path, "r") as f:
    X = f["tracings"][:]  

print("EEG data shape:", X.shape)


labels = pd.read_csv(csv_path).iloc[:, 0].to_numpy()

print(labels)
```

```python
# ---------- Step 4: Fix mismatch (3600 vs 3599) ----------
min_len = min(len(X), len(labels))
X = X[:min_len]
y = labels[:min_len]
print("After trimming → X:", X.shape, "y:", y.shape)

```

```python

# Convert (1280, 20) → (20, 1280)
X = np.transpose(X, (0, 2, 1))  # Now shape: (N, 20, 1280)
print("Final EEG shape (ready for model):", X.shape)


```

```python
# Merge the dataset checking
dataset = [(X[i], y[i]) for i in range(len(y))]


#create one sample data and try to see.
sample_x, sample_y = dataset[0]
print("Example sample shape:", sample_x.shape, "| Label:", sample_y)

```

```python
#finally marge the dataset and save it 
np.savez("merged_eeg_dataset.npz", X=X, y=y)
print("Merge Complete")
```

```python
data = np.load("merged_eeg_dataset.npz")
X, y = data["X"], data["y"]
```
